# Speed vs memory usage

This code will generate a line chart showing how allocator speed performance is affected with memory usage.

## Parameters

```python
input_csv = 'data/speed_benchmarks.csv'

op = 'Allocate'
mem_size = 134217728
bl1_ratio = 50
use_log = True

use_log_text = 'log ' if use_log else ''
title = f"{op} performance - speed vs memory usage"
file = f'rendered/{op} # speed vs memory usage {use_log_text}# memsize {mem_size} # b1 {bl1_ratio}.png'

save = False
```

## The code

### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator, ScalarFormatter, LogLocator)
```

### Load & filter data

```python
all_benchmarks = pd.read_csv(input_csv)

data = all_benchmarks[
    (all_benchmarks["Operation"] == op) &
    (all_benchmarks["Memory size"] == mem_size) &
    (all_benchmarks["Block 1 ratio"] == bl1_ratio)].groupby('Allocator')

print(len(data))
```

### Generate values on x axis

```python
data_xticks = all_benchmarks[all_benchmarks["Operation"] == op]['Memory usage'].unique()
print(data_xticks)
```

### Plot

```python
plt.style.use('dark_background')
plt.rcParams['font.family'] = 'Century Gothic'
plt.figure(figsize=(10, 6))
plt.title(title, fontsize=16, pad=20)
#plt.gca().invert_yaxis()
ax = plt.subplot()

label_offset = 1

for group in data:
    myplot = ax.plot(group[1]['Memory usage'],
                     group[1]['Average (us)'],
                     '-o',
                     label=group[0]
                     )

    max_mem_size = group[1]['Memory usage'].max()
    max_row = group[1][group[1]['Memory usage'] == max_mem_size]

    #
    # Fine tune y offset of labels, so they can be read.
    # These probably need to be adjusted if the data changes
    #
    label_yoffset = 0
    if 'Buddy' in group[0]:
        label_yoffset=1
    elif 'Bitmap Allocator (First Fit)' in group[0]:
        label_yoffset=-2
    elif 'Linked List Allocator (Worst Fit)' in group[0]:
        label_yoffset=-20
    elif 'Bitmap Allocator (Worst Fit)' in group[0]:
        label_yoffset=30
    
    
    
    plt.text(label_offset+max_row['Memory usage'].iloc[0],
             label_yoffset + max_row['Average (us)'].iloc[0],
             group[0],
             va = 'center',
             c = myplot[0].get_color())
          #ax.plot(data['Allocator'],
          #data['Average (us)'],
          #color='#C678DD',
          #xerr=error,
          #ecolor='#ccc',
          #alpha=0.7,
          #capsize=4,
          #height=.7)

ax.set_xlabel('Memory usage (%)')
ax.set_xticks(data_xticks)
ax.set_ylabel('Time (ns) - lower is better')
if use_log:
    ax.set_yscale('log')
ax.yaxis.set_major_formatter(ScalarFormatter())
if use_log:
    ax.yaxis.set_major_locator(LogLocator(base=10, subs=(0.25,0.5,1.0), numticks=20))
#ax.xaxis.set_minor_locator(MultipleLocator(10))
ax.grid(True, axis='y', color='white', alpha=.1)
ax.spines[['right', 'top']].set_visible(False)
fig = plt.gcf()
fig.tight_layout()
if save:
    fig.savefig(file, transparent=True, dpi=300)
plt.show()
```
